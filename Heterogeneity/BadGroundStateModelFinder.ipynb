```python
path = pathlib.Path('/dls/data2temp01/labxchem/data/2017/lb18145-17/processing/edanalyzer/output/pandda_new_score/panddas_new_score')
```

```python
if pathlib.Path('model_scores.csv').exists():
    df = pd.read_csv('model_scores.csv')
else:
    df = pd.DataFrame(columns=['system', 'dtag', 'model_number', 'res', 'score'])
```

```python
df
```

```python
model_records = []
for pandda_dir in path.glob('*'):

    if pandda_dir.name in df['system'].unique():
        continue
    processed_datasets_dir = pandda_dir / 'processed_datasets'
    if not processed_datasets_dir.exists():
        continue
    # print(pandda_dir.name)
    # rprint(model_records)
    for dataset_dir in processed_datasets_dir.glob('*'):
        processed_dataset_file = dataset_dir / 'processed_dataset.yaml'
        if not processed_dataset_file.exists():
            continue
        with open(processed_dataset_file, 'r') as f:
            processed_dataset = yaml.safe_load(f)
        for model_number in processed_dataset['Models']:
            model_info = processed_dataset['Models'][model_number]
            model_records.append(
                {'system': pandda_dir.name, 
                 'dtag': dataset_dir.name, 
                 'model_number': model_number, 
                 'res': processed_dataset['Summary']['Processing Resolution'],
                 'score': model_info['Model Score']
                }
            )

    df = pd.DataFrame(model_records)
    df.to_csv('model_scores.csv')
    print(len(df))
    print(df.tail(1))
```
